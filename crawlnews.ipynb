```python
from bs4 import BeautifulSoup
from urllib.request import urlopen


```

```python
url='http://www.khan.co.kr/'
webpage = urlopen(url)
source = BeautifulSoup(webpage, 'lxml')
titles=source.select('div[id=many_list]')
```

```python
print(source)
```

```python
for title in titles:
    print(title.get_text())
```
